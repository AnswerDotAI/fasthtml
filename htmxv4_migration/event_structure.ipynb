# Click-Load Demo (htmx v4)

Migrated from `click-load.py` (htmx v2).

- `PicoBusy()` → `PicoBusy(htmx4=True)`
    - Event names changed from `beforeRequest` / `afterRequest` to `before-request` / `after-request` in FastHTML with `metaCharacter="-"`
    - Event detail path changed from `event.detail.elt` to `event.detail.ctx.sourceElement`


```python
from fasthtml.common import *
from fasthtml.jupyter import *
from fasthtml.pico import PicoBusy
```

```python
from asyncio import sleep
from random import randint
import secrets

app,rt = fast_app(htmx4=True)

def mk_row():
    idx = randint(0,1000)
    return Tr( Td(f'Agent {idx}'), Td(f'void{idx}@null.org'), Td(secrets.token_hex(8)))

@rt('/')
def get():
    return Titled('Click load demo',
        PicoBusy(htmx4=True),
        P(I('Loading is delayed by 1 second to demonstrate use of "aria-busy" loading spinner in Pico.')),
        Table(
            Thead( Tr( Th('Name'), Th('Email'), Th('ID'))),
            Tbody( *[mk_row() for _ in range(5)], id='agents')),
        Button('Load Another Agent...',
               hx_get='/more', hx_target='#agents', hx_swap='beforeend', cls='btn primary')
    )

@rt('/more')
async def get():
    await sleep(1)
    return mk_row()

srv = JupyUvi(app)
```
