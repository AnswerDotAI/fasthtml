# ws_no4 — Simple click-to-post (htmx v4)

### Migration changes from v2 to v4

| Change | v2 | v4 |
|--------|----|----|
| App setup | `fast_app()` | `fast_app(htmx4=True)` |
| Request header | `HX-Trigger` (just id: `textid`) | `HX-Source` (tag#id format: `p#textid`) |
| Handler param | `hx_trigger: str` | `hx_source: str` |

```python
from fasthtml.common import *
from fasthtml.jupyter import *
```

```python
app, rt = fast_app(htmx4=True)

@rt
def index():
    return Div(
        P(id="textid", name="textname", hx_post=handle, hx_target='#dest')(
            "Click Me"),
        P("Text to Update", id="dest"))

@rt
async def handle(hx_source:str): return Div(f"source: {hx_source}")

srv = JupyUvi(app)
```
