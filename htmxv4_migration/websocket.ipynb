Migrated from `ws.py` (htmx v2).

**Changes from v2 → v4:**
- `hx_ext="ws"`: removed (extension auto-registers in v4)
- `ws_connect="/ws"` → `hx_ws_connect="/ws"`
- `ws_send=True` → `hx_ws_send=True`
- `hx_swap_oob=True` (must be explicit in ws htmxv4)

```python
from fasthtml.common import *
from fasthtml.jupyter import *
```

```python
app, rt = fast_app(exts='ws', htmx4=True)

@rt("/")
def get():
    return Div(
        P(hx_ws_send=True, id="textid", name="textname", hx_ws_connect="/ws")(
            "Click Me"),
        P("Text to Update", id="dest")
    )

@app.ws("/ws")
async def ws(ws, send, hx_source:str):
    return Div(f"trigger: {hx_source}", id="dest", hx_swap_oob=True)

srv = JupyUvi(app)
```
