# Attribute Inheritance Demo (htmx v4)

htmx v4 disables implicit inheritance by default.
To let child elements inherit an attribute from a parent, use the `:inherited` modifier.

With FastHTML's `metaCharacter="-"`, this becomes a Python keyword like `hx_target_inherited`.


```python
from fasthtml.common import *
from fasthtml.jupyter import *

app, rt = fast_app(htmx4=True)

@rt('/')
def get():
    return Titled(
        'Attribute Inheritance Demo',
        P('Both buttons inherit `hx-target` from their parent.'),
        Div(
            Button('Load A', hx_get='/a', cls='btn primary'),
            Button('Load B', hx_get='/b', cls='btn secondary'),
            hx_target_inherited='#output', # child buttons inherit #output as their hx_target
            cls='flex gap-2'
        ),
        Div(P('Waiting...'), id='output')
    )

@rt('/a')
def a():
    return P('Loaded A')

@rt('/b')
def b():
    return P('Loaded B')

srv = JupyUvi(app)
```
